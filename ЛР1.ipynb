```python
import matplotlib.pyplot as plt
import numpy as nm
import math

par1 = float(input("Введіть перший параметр:"))
par2 = float(input("Введіть другий параметр:"))
x = nm.arange(0,1,0.01)
y0 = 0
y = nm.empty(x.size)
y1 = nm.empty(x.size)
y2 = nm.empty(x.size)

y3 = nm.empty(x.size)
y4 = nm.empty(x.size)
y5 = nm.empty(x.size)
for xt in x:
                y[y0] = 1. - (math.e**(-par1*xt))
                y1[y0] = 1- y[y0]
                y2[y0] = par1 * (math.e**(-par1*xt))
                
                y3[y0] = 1. - (math.e**(-par2*xt))
                y4[y0] = 1- y3[y0]
                y5[y0] = par2 * (math.e**(-par2*xt))
               
                y0= y0+1
                
plt.plot(x,y,'r',x,y1,'g',x,y2,'b')
plt.axis([0,1,0,par1])
plt.legend(['P(t)','Q(t)','f(t)'])
plt.show()

plt.plot(x,y3,'r',x,y4,'g',x,y5,'b')
plt.axis([0,1,0,par2])
plt.legend(['P(t)','Q(t)','f(t)'])
plt.show()
```
